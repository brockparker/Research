```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Mon Aug  5 10:49:46 2024

@author: baparker
"""

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import astropy.constants as c
import astropy.units as u
# BP Neccessary imports.

plt.rc('axes', labelsize=14)
plt.rc('figure', titlesize=30)
plt.rc('xtick', labelsize=12)
plt.rc('ytick', labelsize=12)
plt.rc('legend', fontsize=12)
# BP Stylization parameters.

compiled_data_file = '/home/baparker/GitHub/Research/CCDs/Sophia/Data/QE/compiled_data.csv'
compiled_data = pd.DataFrame(pd.read_csv(compiled_data_file))

print(compiled_data)

fig, ax = plt.subplots(layout='tight', figsize=(6,4))

wavelengths = np.array(compiled_data['wavelength']) * u.nm
qe = np.array(compiled_data['qe'])

qe /= qe.max()

energy = (c.h * c.c / wavelengths).to(u.eV)

ax.plot(energy, qe)

ax.set_xlabel('Photon Energy [eV]')
ax.set_ylabel('Relative QE')
ax.set_title('Sophia QE')
ax.tick_params(axis='both', direction='in', which='both')
ax.set_xlim(1, 1e5)
ax.set_xscale('log')
fig.tight_layout()
plt.savefig('/home/baparker/GitHub/Research/CCDs/Sophia/Data/QE/sophia_qe.png', dpi = 500)
plt.show()
```
