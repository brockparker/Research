```python
def linear(x, m, b):
    return m*x + b

x = np.arange(1, 1000, 1)

line1 = linear(x, 1, 10)
line2 = linear(x, 18, 2)
```

```python
plt.plot(x, line1)
plt.plot(x, line2)
plt.plot(x, line1 + line2)
```

```python
def knee(x, m1, m2, b1, b2, k1, k2):
    def line(x, m, b):
        return m * x + b
    weight = np.arctan(k1 * x - k2)/np.pi + 0.5
    plt.plot(weight)
    return line(x, m1, b1) * weight + line(x, m2, b2) * (1- weight)
```

```python
x = np.arange(1, 200, 1)
plt.plot(x, knee(x, 0.1, 0, 0, 0.1, 0.2, 20))
```
