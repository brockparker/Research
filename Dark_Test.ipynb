```python
def knee(x, m1, m2, b1, b2, k1, k2):
    def line(x, m, b):
        return m * x + b
    weight = np.arctan(k1 * x - k2)/np.pi + 0.5
    plt.plot(weight)
    return line(x, m1, b1) * weight + line(x, m2, b2) * (1- weight)
```

```python
x = np.arange(1, 300, 1)
plt.plot(x, knee(x, 2, 0, 0, 100, 0.2, 20))

print(20 * 2 * np.pi)
```
